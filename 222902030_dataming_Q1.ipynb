```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/rakesh22230/Data-Mining-Lab/refs/heads/main/Bank_Credit_Card_Dataset.csv")
print(df.head(5))
```

```python
df= df[ ['Credit_Limit','Avg_Open_To_Buy','Customer_Age']]
df
```

```python
df.isnull().sum()
```

```python
df['Credit_Limit'] = df['Credit_Limit'].fillna(df['Credit_Limit'].mean())
df['Customer_Age'] = df['Customer_Age'].fillna(df['Customer_Age'].mean())
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
from scipy.cluster.hierarchy import linkage

# Perform hierarchical clustering using Ward's method
linked = linkage(scaled_data, method='ward')
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(10, 6))
dendrogram(linked)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```
