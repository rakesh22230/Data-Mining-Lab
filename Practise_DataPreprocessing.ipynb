```python
import pandas as pd
import numpy as np

data = {
    "Item": ["Rice","Oil","Lentil","Sugar","Rice","Oil","Salt","Rice","Lentil","Oil"],
    "Category": ["Food","Oil","Food","Food","Food","Oil","Spice","Food","Food","Oil"],
    "Quality": ["High","Medium","Low","Medium","High","High","Low","Medium","Low","High"],
    "Price": [100, 850, 150, np.nan, 110, 900, 40, 95, np.nan, 880],
    "Quantity_KG": [5, np.nan, 2, 1, 10, 8, 1, np.nan, 3, 7],
    "Discount": [2, 50, 1, 1000, 3, 60, 0, 2, 1, 55],  # noisy
    "Supplier": ["A","B","A","C","A","B","C","A","A","B"],
    "Sold": ["Yes","Yes","No","Yes","Yes","Yes","No","Yes","No","Yes"]
}

df = pd.DataFrame(data)
df
```

```python
if 'Supplier' in df.columns:
    df.drop(["Supplier"], axis=1, inplace=True)
df
```

```python
df.info()
df.isnull().sum()

```

```python
df["Price"] = df["Price"].fillna(df["Price"].mean())
df["Quantity_KG"] = df["Quantity_KG"].fillna(df["Quantity_KG"].median())

df
```

```python
df["Discount_Level"] = pd.cut(
    df["Discount"],
    bins=[-1, 5, 50, 2000],
    labels=["Low","Medium","High"]
)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Sold"] = le.fit_transform(df["Sold"])

df
```
