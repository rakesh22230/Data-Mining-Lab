Linear Regression for Bottle_CSV


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/content/sample_data/bottle.csv")
print(df.head(3))
print("\n\n")
df_extracted = df[ ['Salnty','T_degC']]
df_extracted.head()
```

```python
sns.scatterplot(x='Salnty', y='T_degC', data=df_extracted )
```

```python
sns.lmplot(x='Salnty', y='T_degC', data=df_extracted,order=2)
```

```python
sns.lmplot(x='Salnty', y='T_degC', data=df_extracted,order=1)
```

```python
df_extracted.fillna(method='ffill', inplace=True)
```

```python
x = np.array(df_extracted['Salnty']).reshape(-1,1)
y = np.array(df_extracted['T_degC']).reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
lin_reg.score(x_test,y_test)
```
